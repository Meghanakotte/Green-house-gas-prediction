```python

import pandas as pd  # Data manipulation and analysis
import numpy as np   # Numerical operations

import seaborn as sns      # Advanced statistical visualizations
import matplotlib.pyplot as plt  # Basic plotting functions

from sklearn.model_selection import train_test_split, GridSearchCV   # Data splitting & hyperparameter tuning
from sklearn.preprocessing import LabelEncoder, StandardScaler       # Encoding categorical features & feature scaling
from sklearn.ensemble import RandomForestRegressor                   # Regression model that handles mixed data types
from sklearn.metrics import mean_squared_error, r2_score             # Model evaluation metrics

import joblib   # Save and load trained models efficiently
import os                              
```

STEP2: LOAD DATASET

```python
excel_file='SupplyChainEmissionFactorsforUSIndustriesCommodities.xlsx'
years=range(2010, 2017)
years[0]
```

```python
# Viewing the dataset
df_1 = pd.read_excel(excel_file, sheet_name=f'{years[0]}_Detail_Commodity')
df_1.head()
```

```python
# Viewing the dataset
df_2 = pd.read_excel(excel_file, sheet_name=f'{years[0]}_Detail_Industry')
df_2.head()
```

```python
all_data = []  # List to store combined data from all years

for year in years:
    try:
        # Read both commodity and industry sheets for the given year
        df_com = pd.read_excel(excel_file, sheet_name=f'{year}_Detail_Commodity')
        df_ind = pd.read_excel(excel_file, sheet_name=f'{year}_Detail_Industry')
        
        # Adding a column to identify source
        df_com['Source'] = 'Commodity'
        df_ind['Source'] = 'Industry'

        # Adding year column to both
        df_com['Year'] = df_ind['Year'] = year
        
        # Remove any extra spaces from column names
        df_com.columns = df_com.columns.str.strip()
        df_ind.columns = df_ind.columns.str.strip()

        # Renameing common columns for consistency
        df_com.rename(columns={
            'Commodity Code': 'Code',
            'Commodity Name': 'Name'
        }, inplace=True)
        
        df_ind.rename(columns={
            'Industry Code': 'Code',
            'Industry Name': 'Name'
        }, inplace=True)
        
        # Combining commodity and industry data and store it
        all_data.append(pd.concat([df_com, df_ind], ignore_index=True))
        
    except Exception as e:
        print(f"Error processing year {year}: {e}")
```

```python
# Combining the excel sheets
df=pd.concat(all_data, ignore_index=True)
df.head()
```

```python
X.dtypes
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

STEP 6: Training and Testing

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

Selecting model for training

```python
model = RandomForestRegressor(random_state=42)
```

```python
model.fit(X_train, y_train)
```

Prediction and Evaluation

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'RMSE: {rmse}')
print(f'R Score: {r2}')
# r2_sccore approx 1 gives us perfecr model
```

Hyperparameter tuning

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}
grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3, n_jobs=-1)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)

```

```python
y_pred_best = best_model.predict(X_test)

print("Final RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_best)))
print("Final R:", r2_score(y_test, y_pred_best))
```

The model before hypertuning is giving better results considering it as final model.
